```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

df = pd.read_csv("lab4\data.csv")
```

```python
plt.figure(figsize=(14,7))
sns.boxplot(
    data=df,
    x="nazwa rozwiazania",
    y="sredni czas oczekiwania",
    hue="liczba filozofow"
)
plt.title("Porwnanie rednich czasw oczekiwania filozofw dla rnych rozwiza")
plt.xlabel("Rozwizanie")
plt.ylabel("redni czas oczekiwania [ms]")
plt.legend(title="Liczba filozofw")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
df["rozwiazanie_skrot"] = df["nazwa rozwiazania"].replace({
    "Rozwizanie z moliwoci zagodzenia": "Z zagodzeniem",
    "Rozwizanie asymetryczne": "Asymetryczne",
    "Rozwizanie z arbitrem": "Z arbitrem",
})
```

```python
g = sns.FacetGrid(df, col="liczba filozofow", col_wrap=2, height=5, sharey=True)
g.map_dataframe(
    sns.boxplot,
    x="rozwiazanie_skrot",
    y="sredni czas oczekiwania",
    hue="rozwiazanie_skrot",
    palette="Set2"
)
g.set_titles("Liczba filozofw: {col_name}")

unique_labels = df["rozwiazanie_skrot"].unique()
palette = sns.color_palette("Set2")
handles = [Patch(facecolor=palette[i], label=label) for i, label in enumerate(unique_labels)]

g.figure.legend(handles=handles, title="Rozwizanie", loc="upper right")

g.set_axis_labels("Rozwizanie", "redni czas oczekiwania [ms]")
plt.tight_layout()
plt.show()
```

```python
df_mean = df.groupby(["rozwiazanie_skrot", "liczba filozofow"])["sredni czas oczekiwania"].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(
    data=df_mean,
    x="liczba filozofow",
    y="sredni czas oczekiwania",
    hue="rozwiazanie_skrot",
    marker="o"
)
plt.title("redni czas oczekiwania w zalenoci od liczby filozofw i rozwizania")
plt.xlabel("Liczba filozofw")
plt.ylabel("redni czas oczekiwania [ms]")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(title="Rozwizanie")
plt.show()
```
